# EDA

```python
df.isnull().sum()
```

```python
[i for i in df.columns if df[i].isnull().sum()>0]
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.isnull(),yticklabels=False)
plt.show()
```

```python
df_country=pd.read_excel("Country-code.xlsx")
df_country.head()
```

```python
df=pd.merge(df,df_country,on='Country Code',how='left')
df.head()
```

```python
df.columns
```

```python
df['Country'].value_counts()
```

```python
country_names=df['Country'].value_counts().index
country_val=df['Country'].value_counts().values
print(country_names)
print(country_val)
```

```python
plt.figure(figsize=(10,8))
plt.pie(country_val[:3], labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
#lets convert above table into a df
ratings=df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
plt.figure(figsize=(12,8))
#plt.style.use('dark_background')
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','olive'])
plt.show()
```

```python
sns.countplot(x='Rating color',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','olive'])
plt.ylabel('Ratings')
```

```python
df[df['Rating color']=='White'].groupby('Country').size().reset_index()
```

```python
#which currency is iused by which country
df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
# which countries has online delivery 
df[['Country','Has Online delivery']].groupby(['Country','Has Online delivery']).size().reset_index()
```

```python
# create a pie chart for city distribution
city_names= df.City.value_counts().index
city_val=df.City.value_counts().values
print(city_names)
print(city_val)
```

```python
plt.figure(figsize=(10,8))
plt.pie(city_val[:5],labels=city_names[:5],autopct='%1.2f%%')
plt.show()
```
